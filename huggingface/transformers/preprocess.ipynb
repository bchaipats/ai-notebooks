```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
```

```python
from datasets import load_dataset, Audio

dataset = load_dataset("PolyAI/minds14", name="en-US", split="train")
```

```python
dataset[0]["audio"]
```

```python
dataset = dataset.cast_column("audio", Audio(sampling_rate=16_000))
```

```python
from transformers import AutoFeatureExtractor

feature_extractor = AutoFeatureExtractor.from_pretrained("facebook/wav2vec2-base")
```

```python
audio_input = [dataset[0]["audio"]["array"]]
feature_extractor(audio_input, sampling_rate=16000)
```

```python
dataset[0]['audio']['array'].shape
```

```python
dataset[1]['audio']['array'].shape
```

```python
def preprocess_function(examples):
    audio_arrays = [x['array'] for x in examples['audio']]
    inputs = feature_extractor(
        audio_arrays,
        sampling_rate=16000,
        padding=True,
        max_length=100000,
        truncation=True,        
    )
    return inputs
```

```python
processed_dataset = preprocess_function(dataset[:5])
```

```python
from datasets import load_dataset

dataset = load_dataset("food101", split="train[:100]")
```

```python
dataset[0]["image"]
```

```python
from transformers import AutoFeatureExtractor

feature_extractor = AutoFeatureExtractor.from_pretrained("google/vit-base-patch16-224")
```

```python
tuple(feature_extractor.size.values())
```

```python
from torchvision.transforms import Compose, RandomResizedCrop, Normalize, ColorJitter, ToTensor

normalize = Normalize(mean=feature_extractor.image_mean, std=feature_extractor.image_std)
_transforms = Compose(
    [
        RandomResizedCrop(size=tuple(feature_extractor.size.values())), 
        ColorJitter(brightness=0.5, hue=0.5), 
        ToTensor(), 
        normalize]
)
```

```python
def transforms(examples):
    examples["pixel_values"] = [_transforms(image.convert("RGB")) for image in examples["image"]]
    return examples
```

```python
dataset.set_transform(transforms)
```

```python
from datasets import load_dataset

lj_speech = load_dataset("lj_speech", split="train")
```

```python
lj_speech = lj_speech.map(remove_columns=["file", "id", "normalized_text"])
```

```python
lj_speech[0]["audio"]
```

```python
lj_speech[0]["text"]
```

```python
lj_speech = lj_speech.cast_column("audio", Audio(sampling_rate=16_000))
```

```python
from transformers import AutoProcessor

processor = AutoProcessor.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
def prepare_dataset(example):
    audio = example["audio"]
    example.update(processor(audio=audio["array"],
                             text=example["text"],
                             sampling_rate=16000))
    return example
```

```python
prepare_dataset(lj_speech[0])
```
